#Revisão de Python - Numpy arrays


Um array numpy é um conjunto de valores, todos do mesmo tipo, e indexado por um valor inteiro.  O tamanho (shape) do array é definido pelo número de tuplas e suas dimensões.


```python
#@title Bibliotecas
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

Criando uma matriz usando a biblioteca Numpy

```python
#criando um array simples
b = np.array([[1,2,3],[4,5,6]])
print(b.shape)
```

Criando uma matriz com valores a aleatórios

```python
#criando um array randômico
e = np.random.randint(5, size=(5, 10)) #((5,10)) 5 linhas e 10 colunas
print(e)
```

Imprimindo valores específicos da matriz

```python
print(e[0:5,0]) #slice
```

Exemplo: Como podemos imprimir todos os elementos da terceira coluna?

Filtrando valores da matriz

```python
e[e>3] # filter
```

Utilizando a operação de soma de linhas (axis=0) ou colunas (axis=1).

```python
#math
np.sum([[0, 1], [2, 3],[4,5]], axis=0)
```

```python
np.min(e)
```

Removendo valores do numpy

    np.delete(arr, obj, axis=None)
    arr: array
    obj: linha ou coluna a remover
    axis: dimensão a apagar

```python
print(b)
np.delete(b,1,1)
```

Removendo valores Nan

```python

a = np.array([[1, 2, 3], [0, 3, np.nan]])
print(a)
where_are_NaNs = np.isnan(a)
a[where_are_NaNs] = 0
a
```

```python


df_covid.plot(x ='date', y='cases')

#‘area’
#'pie'
```

Como comparar o número de mortes entre RS e RJ?

```python
df_rs=df_covid[df_covid['state']=='RS']
df_rj=df_covid[df_covid['state']=='RJ']
plt.plot( 'date', 'deaths', data=df_rs, marker='o', markerfacecolor='yellow',label="RS")
plt.plot( 'date', 'deaths', data=df_rj, marker='_', markerfacecolor='blue',label="RJ")
plt.legend()
#plt.yscale("log")
plt.show()
```

```python
#Mas e se diluirmos o gráfico de acordo com a população de cada estado

#população
pop_rj=17366189
pop_rs=11422973

#mortes por 100k
df_prop_rs=(df_rs['deaths']*100000).div(pop_rs)
df_prop_rj=(df_rj['deaths']*100000).div(pop_rj)

plt.plot( df_rs['date'], df_prop_rs,  marker='o', markerfacecolor='red',label="RS")
plt.plot( df_rj['date'], df_prop_rj,  marker='_', markerfacecolor='blue',label="RJ")
plt.legend()
#plt.yscale("log")
plt.show()
```
